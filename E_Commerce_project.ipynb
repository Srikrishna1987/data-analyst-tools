```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding ='latin-1')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

converting date columns

```python
data['order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] =data['order Date'].dt.month
data['Order Year'] =data['order Date'].dt.year
data['Order Day of Week'] =data['order Date'].dt.dayofweek
```

```python
data.head()
```

# **Monthly sales analysis**

```python
import plotly.express as px

sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
fig =px.line(sales_by_month,x='Order Month',y='Sales',title='Monthly Sales Analysis')
fig.show()
```

```python
data.head()
```

# **Sales by Catagory**

```python
sales_by_catagory = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_catagory
```

```python
fig = px.pie(sales_by_catagory,values='Sales',names='Category',hole=0.5,color_discrete_sequence=colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo='percent+label')
fig.update_layout(title_text='Sales by Catagory',title_font=dict(size=24))
fig.show()
```

# **Sales analysis by Sub Catagory**

```python
data.head()
```

```python
sales_by_subcatagory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcatagory
```

```python
fig =px.bar(sales_by_subcatagory,x='Sub-Category',y='Sales',color='Sub-Category',title='Sales Analysis by Sub Catagory')
fig.show()
```

# **Monthly profit analysis**

```python
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.line(profit_by_month,x='Order Month',y='Profit',title='Monthly Profit Analysis')
fig.show()
```

# **profit by catagory**

```python
profit_by_catagory = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_catagory
```

```python
fig = px.pie(profit_by_catagory,values='Profit',names='Category',hole=0.5,color_discrete_sequence=colors.qualitative.Pastel)
fig.update_traces(textposition='inside',textinfo='percent+label')
fig .update_layout(title_text='Profit analysis by Catagory',title_font=dict(size=24))
fig.show()
```

# **profit by subcatagory**

```python
profit_by_catagory  = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_catagory,x='Sub-Category',y='Profit',color='Sub-Category',title='Profit Analysis by Sub Catagory')
fig.show()
```

# **Sales and Profit -Customer Segment**

```python
data.head()
```

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
color_pallete = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(name='Profit',x=sales_profit_by_segment['Segment'],y=sales_profit_by_segment['Profit'],marker_color=color_pallete[0]))
fig.add_trace(go.Bar(name='Sales',x=sales_profit_by_segment['Segment'],y=sales_profit_by_segment['Sales'],marker_color=color_pallete[1]))
fig.update_layout(title='Sales and Profit by Customer Segment',
                  xaxis_title= 'Customer Segment',
                  yaxis_title='Amount')

fig.show()
```

# **Sales to profit ratio**

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales to Profit Ratio'] = sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales to Profit Ratio']])
```

Based on the analysis:

*   **Monthly Sales and Profit:** Sales and profit show a general upward trend throughout the year, with peaks typically occurring in the later months (September, October, November, and December). This suggests that the business experiences seasonality, with higher activity towards the end of the year.
*   **Sales and Profit by Category:** Technology products contribute the most to both sales and profit, followed by Office Supplies and then Furniture. This indicates that technology is a strong performing category for this business.
*   **Sales and Profit by Sub-Category:** The analysis by sub-category revealed more detailed insights. Some sub-categories within the Furniture and Office Supplies categories, such as 'Bookcases', 'Tables', and 'Supplies', appear to be generating losses. This highlights areas that require further investigation and potential strategic adjustments. Conversely, 'Phones', 'Chairs', and 'Storage' sub-categories show high sales and profit.
*   **Sales and Profit by Customer Segment:** The 'Consumer' segment contributes the highest amount to both sales and profit, followed by 'Corporate' and 'Home Office'. This suggests that the consumer market is a significant driver of revenue and profitability.
*   **Sales to Profit Ratio:** The 'Home Office' segment has the lowest sales to profit ratio, indicating that for every dollar of sales, they generate a higher profit compared to the other segments. This could be an area to focus on for potentially higher profitability per sale.
